```python
import math
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
sigmoid = 1 / (1 + sp.exp(-x))
sigmoid_func = sp.lambdify(x, sigmoid, 'numpy')

x_vals = np.linspace(-10, 10, 400)
y_vals = sigmoid_func(x_vals)

plt.plot(x_vals, y_vals, label='Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function Plot')
plt.legend()
plt.grid(True)
plt.show()
```
